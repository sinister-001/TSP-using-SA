```python
import math
import random
import matplotlib.pyplot as plt

# Define the distance function between two points
def distance(point1, point2):
    x1, y1 = point1
    x2, y2 = point2
    return math.sqrt((x1 - x2) ** 2 + (y1 - y2) ** 2)

# Define the total distance function for a tour
def total_distance(points, tour):
    dist = 0
    for i in range(len(tour)):
        j = (i + 1) % len(tour)
        dist += distance(points[tour[i]], points[tour[j]])
    return dist

# Generate random points for cities
random.seed(42)
points = [(random.uniform(0, 10), random.uniform(0, 10)) for _ in range(15)]

# Plot the points
fig, ax = plt.subplots()
ax.scatter([p[0] for p in points], [p[1] for p in points])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('15 Cities')

# Randomly choose a starting city
start_city = random.randint(0, len(points) - 1)

# Define the initial tour as visiting all cities in order
tour = list(range(len(points)))
tour.remove(start_city)
tour = [start_city] + tour

# Plot the initial tour
x = [points[i][0] for i in tour + [tour[0]]]
y = [points[i][1] for i in tour + [tour[0]]]
ax.plot(x, y, color='r')

# Define the simulated annealing algorithm
def simulated_annealing(points, tour, start_temp, cooling_factor):
    # Initialize the current and best tours
    current_tour = tour
    best_tour = tour
    # Calculate the initial cost
    current_cost = total_distance(points, current_tour)
    best_cost = current_cost
    # Initialize the temperature
    temp = start_temp
    # Initialize the iteration counter
    iteration = 0
    # Initialize lists to store the iteration and cost values
    iterations = []
    costs = []
    # Run the simulated annealing algorithm
    while temp > 1e-10:
        # Generate a random neighbor by swapping two cities
        i, j = sorted(random.sample(range(len(points)), 2))
        new_tour = current_tour[:i] + current_tour[j:j+1] + current_tour[i+1:j] + current_tour[i:i+1] + current_tour[j+1:]
        # Calculate the new cost
        new_cost = total_distance(points, new_tour)
        # Determine whether to accept the new tour
        if math.exp((current_cost - new_cost) / temp) > random.uniform(0, 2):
            current_tour = new_tour
            current_cost = new_cost
        # Update the best tour and cost if necessary
        if current_cost < best_cost:
            best_tour = current_tour
            best_cost = current_cost
        # Reduce the temperature
        temp *= cooling_factor
        # Increment the iteration counter
        iteration += 1
        # Record the iteration and cost values
        iterations.append(iteration)
        costs.append(best_cost)
    # Return the best tour and cost, and the iteration and cost values
    return best_tour, best_cost, iterations, costs

# Run the simulated annealing algorithm with a starting temperature of 1e+10 and a cooling factor of 0.97
best_tour, best_cost, iterations, costs = simulated_annealing(points, tour, 1e+10, 0.97)

# Plot the iteration vs cost graph
fig, ax = plt.subplots()
ax.plot(iterations, costs)
ax.set_xlabel('Iteration')
ax.set_ylabel('Cost')
ax.set_title('SA Optimization')

# Plot the final tour
x = [points[i][0] for i in best_tour + [best_tour[0]]]
y = [points[i][1] for i in best_tour + [best_tour[0]]]
fig, ax = plt.subplots()
ax.scatter([p[0] for p in points], [p[1] for p in points])
ax.plot(x, y, color='r')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Final Tour')

# Print the best tour and cost
print(f'Best Tour: {best_tour}')
print(f'Best Cost: {best_cost}')
```

```python
from math import inf
from copy import deepcopy

# Tic Tac Toe board
board = [['-', '-', '-'],
         ['-', '-', '-'],
         ['-', '-', '-']]

# Constants
HUMAN = 'X'
AI = 'O'
EMPTY = '-'

# Alpha-Beta pruning function
def minimax(state, depth, alpha, beta, player):
    # Check for terminal state
    winner = check_winner(state)
    if winner == AI:
        return (10 - depth, None)
    elif winner == HUMAN:
        return (-10 + depth, None)
    elif winner == 'Tie':
        return (0, None)

    # Maximizing player's turn
    if player == AI:
        max_val = -inf
        move = None

        for i in range(3):
            for j in range(3):
                if state[i][j] == EMPTY:
                    state[i][j] = AI
                    score, _ = minimax(state, depth + 1, alpha, beta, HUMAN)
                    state[i][j] = EMPTY
                    if score > max_val:
                        max_val = score
                        move = (i, j)

                    alpha = max(alpha, max_val)
                    if alpha >= beta:
                        break

        return (max_val, move)

    # Minimizing player's turn
    else:
        min_val = inf
        move = None

        for i in range(3):
            for j in range(3):
                if state[i][j] == EMPTY:
                    state[i][j] = HUMAN
                    score, _ = minimax(state, depth + 1, alpha, beta, AI)
                    state[i][j] = EMPTY
                    if score < min_val:
                        min_val = score
                        move = (i, j)
                    beta = min(beta, min_val)
                    if alpha >= beta:
                        break
        return (min_val, move)

# Function to check for winner
def check_winner(state):
    # Check rows
    for row in state:
        if row[0] == row[1] == row[2] and row[0] != EMPTY:
            return row[0]

    # Check columns
    for col in range(3):
        if state[0][col] == state[1][col] == state[2][col] and state[0][col] != EMPTY:
            return state[0][col]

    # Check diagonals
    if state[0][0] == state[1][1] == state[2][2] and state[0][0] != EMPTY:
        return state[0][0]
    if state[0][2] == state[1][1] == state[2][0] and state[0][2] != EMPTY:
        return state[0][2]

    # Check for tie
    for row in state:
        for val in row:
            if val == EMPTY:
                return None

    return 'Tie'

# Function to print the board
def print_board(state):
    print('-------------')
    for i in range(3):
        print('|', end=' ')
        for j in range(3):
            print(state[i][j], '|', end=' ')
        print()
        print('-------------')

# Main function to play the game
def play():
    current_player = HUMAN

    while True:
        print_board(board)

        if current_player == HUMAN:
            print('Your turn:')
            row = int(input('Enter row (1, 2, 3) : '))
        
        while row not in range(1, 4):
            row = int(input('Enter row (1, 2, 3): '))
        col = int(input('Enter column (1, 2, 3): '))
        while col not in range(1, 4):
            col = int(input('Enter column (1, 2, 3): '))

        if board[row - 1][col - 1] != EMPTY:
            print('Invalid move. Try again.')
            continue

        board[row - 1][col - 1] = HUMAN
        current_player = AI

    else:
        print('Computer\'s turn:')
        _, move = minimax(deepcopy(board), 0, -inf, inf, AI)
        board[move[0]][move[1]] = AI
        current_player = HUMAN

    winner = check_winner(board)
    if winner:
        print_board(board)
        if winner == HUMAN:
            print('Congratulations! You win.')
        elif winner == AI:
            print('Sorry, you lose. Better luck next time.')
        else:
            print('It\'s a tie.')
            break

play()
```

```python
from copy import deepcopy

# Constants
EMPTY = 0
HUMAN = 1
AI = 2

# Board representation
board = [[EMPTY, EMPTY, EMPTY],
         [EMPTY, EMPTY, EMPTY],
         [EMPTY, EMPTY, EMPTY]]


# Function to print the board
def print_board(board):
    for row in board:
        print(' '.join(['X' if cell == HUMAN else 'O' if cell == AI else '-' for cell in row]))


# Function to check for a winner
def check_winner(board):
    for player in (HUMAN, AI):
        # Check for horizontal win
        for row in range(3):
            if board[row][0] == board[row][1] == board[row][2] == player:
                return player

        # Check for vertical win
        for col in range(3):
            if board[0][col] == board[1][col] == board[2][col] == player:
                return player

        # Check for diagonal win
        if board[0][0] == board[1][1] == board[2][2] == player:
            return player
        if board[0][2] == board[1][1] == board[2][0] == player:
            return player

    # Check for tie
    for row in range(3):
        for col in range(3):
            if board[row][col] == EMPTY:
                return None
    return 0


# Function to evaluate the board for the AI player
def evaluate(board):
    score = 0
    for player in (HUMAN, AI):
        # Check for horizontal lines
        for row in range(3):
            line = board[row]
            if line.count(player) == 3:
                score += 100
            elif line.count(player) == 2 and line.count(EMPTY) == 1:
                score += 5

        # Check for vertical lines
        for col in range(3):
            line = [board[row][col] for row in range(3)]
            if line.count(player) == 3:
                score += 100
            elif line.count(player) == 2 and line.count(EMPTY) == 1:
                score += 5

        # Check for diagonal lines
        line = [board[i][i] for i in range(3)]
        if line.count(player) == 3:
            score += 100
        elif line.count(player) == 2 and line.count(EMPTY) == 1:
            score += 5

        line = [board[i][2 - i] for i in range(3)]
        if line.count(player) == 3:
            score += 100
        elif line.count(player) == 2 and line.count(EMPTY) == 1:
            score += 5

    return score


# Minimax algorithm with alpha-beta pruning
def minimax(board, depth, alpha, beta, player):
    winner = check_winner(board)
    if winner:
        if winner == AI:
            return 100 - depth, None
        elif winner == HUMAN:
            return depth - 100, None
        else:
            return 0, None

    if player == AI:
        max_eval = -float('inf')
        best_move = None
        for row in range(3):
            for col in range(3):
                if board[row][col] == EMPTY:
                    board[row][col] = AI
                    # Minimax algorithm with alpha-beta pruning (continued)
                    eval, _ = minimax(board, depth + 1, alpha, beta, HUMAN)
                    board[row][col] = EMPTY
                    if eval > max_eval:
                        max_eval = eval
                        best_move = (row, col)
                    alpha = max(alpha, eval)
                    if beta <= alpha:
                        break
        return max_eval, best_move

    else:
        min_eval = float('inf')
        best_move = None
        for row in range(3):
            for col in range(3):
                if board[row][col] == EMPTY:
                    board[row][col] = HUMAN
                    eval, _ = minimax(board, depth + 1, alpha, beta, AI)
                    board[row][col] = EMPTY
                    if eval < min_eval:
                        min_eval = eval
                        best_move = (row, col)
                    beta = min(beta, eval)
                    if beta <= alpha:
                        break
        return min_eval, best_move


# Function to play the game
def play():
    current_player = HUMAN
    while True:
        print_board(board)
        winner = check_winner(board)
        if winner:
            if winner == HUMAN:
                print('Congratulations! You win.')
            elif winner == AI:
                print('Sorry, you lose. Better luck next time.')
            else:
                print('It\'s a tie.')
            break

        if current_player == HUMAN:
            print('Your turn:')
            row = int(input('Enter row (1, 2, 3): '))
            while row not in range(1, 4):
                row = int(input('Enter row (1, 2, 3): '))
            col = int(input('Enter column (1, 2, 3): '))
            while col not in range(1, 4):
                col = int(input('Enter column (1, 2, 3): '))

            if board[row - 1][col - 1] != EMPTY:
                print('Invalid move. Try again.')
                continue

            board[row - 1][col - 1] = HUMAN
            current_player = AI

        else:
            print('Computer\'s turn:')
            _, move = minimax(deepcopy(board), 0, -float('inf'), float('inf'), AI)
            board[move[0]][move[1]] = AI
            current_player = HUMAN

# Start the game
play()
```
